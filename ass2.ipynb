```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
```

```python
data = pd.DataFrame({
    'Age': [22, 25, 47, 52, 46, 56, 23, 40, 60, 48],
    'Income': [15000, 29000, 48000, 52000, 46000, 60000, 20000, 35000, 65000, 48000],
    'Student': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes'],
    'Credit_Rating': ['Fair', 'Excellent', 'Fair', 'Excellent', 'Fair', 'Excellent', 'Fair', 'Fair', 'Excellent', 'Fair'],
    'Buys_Computer': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes']
})

print("Original Dataset:\n", data)
```

```python
le = LabelEncoder()
for col in data.columns:
    data[col] = le.fit_transform(data[col])

X = data.drop('Buys_Computer', axis=1)
y = data['Buys_Computer']

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
models = {
    "1. ID3 (Entropy)": DecisionTreeClassifier(criterion='entropy'),
    "2. CART (Gini)": DecisionTreeClassifier(criterion='gini'),
    "3. Max Depth = 3": DecisionTreeClassifier(max_depth=3),
    "4. Min Samples Split = 4": DecisionTreeClassifier(min_samples_split=4),
    "5. Min Samples Leaf = 2": DecisionTreeClassifier(min_samples_leaf=2),
    "6. Pruned Tree (ccp_alpha=0.01)": DecisionTreeClassifier(ccp_alpha=0.01),
    "7. Extra Trees": ExtraTreesClassifier(n_estimators=10, random_state=42)
}
```

```python
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    results.append([name, acc])
    print(f"{name} Accuracy: {acc:.2f}")
```
